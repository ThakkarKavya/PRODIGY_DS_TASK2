```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("C:\\Users\\thakk\\OneDrive\\Desktop\\DATA2.csv")
```

```python
df
```

```python
print("Missing values before cleaning:")
print(df.isnull().sum())
```

```python
if df['Age'].isnull().sum()==0:
    print("No missing values in age column")
```

```python
df['HasCabin']=df['Cabin'].notna().astype(int)
```

```python
df.drop('Cabin',axis=1,inplace=True)
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
print("Missing values After cleaning:")
print(df.isnull().sum())
```

```python
df.describe(include='all')
```

```python
print("Distribution of age")
plt.figure(figsize=(10,6))
sns.histplot(df['Age'],kde=True,bins=30)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
print("Distribution of Fare")
plt.figure(figsize=(10,6))
sns.histplot(df['Fare'],kde=True,bins=30)
plt.title('Distribution of Fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Pclass',y='Survived',data=df)
plt.title('Survival rate by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Survival Rate')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Sex',y='Survived',data=df)
plt.title('Survival rate by Sex')
plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.kdeplot(data=df[df['Survived']==1]['Age'], shade=True,label='Survived',color='orange')
sns.kdeplot(data=df[df['Survived']==0]['Age'], shade=True,label='Not Survived',color='blue')
plt.title('Survival Rate By Age')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
numeric_cols=df.select_dtypes(include=['number'])
```

```python
corr_matrix=numeric_cols.corr()
plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
